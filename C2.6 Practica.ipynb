## Ejercicios sobre una escuela

A lo largo de estos ejercicios utilizaremos el siguiente dataframe:

|name   |school |teacher |sex    | math_score| reading_score|
|:------|:------|:-------|:------|----------:|-------------:|
|mike   |south  |johnson |male   |          4|             1|
|carol  |south  |johnson |female |          3|             5|
|greg   |south  |johnson |male   |          2|             2|
|marcia |south  |johnson |female |          4|             4|
|peter  |north  |smith   |male   |          3|             5|
|jan    |north  |smith   |female |          4|             4|
|bobby  |north  |smith   |male   |          5|             1|
|cindy  |south  |perry   |female |          4|             5|
|alice  |south  |perry   |female |          5|             4|

1. Construye el dataframe anterior y llámalo ``scores``

2. Construye y guarda (es decir, asigna un nombre) al dataframe formado por sus primeras tres filas. Llámalo ``small_scores``

3. Ordena ``scores`` por calificación de matemáticas en orden descendente. ¿Quién tuvo la mejor calificación?

4. Ordena ``scores`` alfabéticamente por nombre

5. Ordena ``scores`` por sexo, de tal manera que primero aparezcan las mujeres. ¿Cuál sexo tiene, en apariencia, mejor calificación de lectura?

6. Ordena ``scores`` por nombre, profesor, sexo, calificación de matemáticas y calificación de lectura.

7. Selecciona únicamente las columnas nombre, calificación de matemáticas y calificación de lectura.

8. Selecciona todas la columnas excepto la de sexo.

9. Selecciona todas las columnas excepto las de calificaciones.

10. Muestra los renglones de los hombres cuya escuela está en el sur.

11. Muestra a los estudiantes que tuvieron buena calificación en matemáticas (tú decides el criterio para decir si una calificación es buena).

12. Muestra a los estudiantes con calificación en matemáticas mayor a 4 y calificación de lectura al menos de 3.

13. Explica los errores de los siguientes códigos:

``scores >> filter(school == south)``

``scores >> filter(school == "south")``

Corrige los códigos anteriores.

14. Te pidieron diseñar un programa de rescate. Muestra los renglones de estudiantes con calificación de a lo más 3 tanto en matemáticas como en lectura.

15. Muestra los renglones cuya calificación de lectura fue 2, 3, o 4.

16. Muestra los renglones cuyo nombre del estudiante comienza con M.

17. Muestra los renglones de los profesores cuyo mejor estudiante de matemáticas obtuvo un 5.

18. ¿Cuál es el sexo cuyo promedio de matemáticas fue 4?

19. Reemplaza los valores de las calificiones por su equivalente en escala de 0 a 50 (5=50, 4=40, etc)

20. Crea una nueva columna donde se indique el promedio de las dos calificaciones de cada alumno.

21. Crea una nueva columna que diga "Sí" si el alumno obtuvo un promedio mayor a 4 en sus dos calificaciones y "No" en caso contrario.

22. Crea una nueva columna donde se muestre la calificación de matemáticas - promedio de la calificación de matemáticas de todos los alumnos.

23. Calcula el promedio de calificaciones en matemáticas por sexo y restalo de las calificaciones de matemáticas. 

24. Muestra el promedio de calificaciones de matemáticas de todos los alumnos.

25. Muestra el promedio de calificaciones de lectura de todos los alumnos.

26. Muestra la máxima y la mínima calificación de matemáticas por escuela.

27. Muestra la máxima calificación de lectura para cada maestro.

28. Si agrupamos por sexo y pedimos la menor calificación de lectura, qué dimensiones tendría el dataframe resultante.

29. Muestra un dataframe donde se indique, para cada sexo, la calificación promedio de lectura así como el total de estudiantes de ese sexo.

30. Selecciona las columnas de nombre y calificación de matemáticas; luego añade una columna donde sumes 5 puntos a cada calificación de matemáticas; luego ordena el dataframe resultante de mayor a menor respecto de esta columna creada.

31. Selecciona todas las columnas excepto la de maestro. Luego añade una columna que muestre el promedio de las dos calificaciones de cada alumno, y ordena el dataframe resultante de menor a mayor respecto de esta columna creada.

32. Muestra la calificación promedio por sexo de los alumnos que cuyo profesor no es Smith.

33. Muestra el máximo, el mínimo y el promedio de las chicas de la escuela del sur.

34. Observa los siguientes códigos:

``

  group_by(_.school, _.teacher) >> 
  summarize(max_math_score = _.math_score.max()) >> 
  filter(_.school == "south")
``

``

  filter(_.school == "south") >> 
  group_by(_.teacher) >>
